```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import six
import joblib
import sys
sys.modules['sklearn.externals.six'] = six
sys.modules['sklearn.externals.joblib'] = joblib
!pip install mglearn
import mglearn

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA, IncrementalPCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, adjusted_rand_score, silhouette_score
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from scipy.cluster.hierarchy import dendrogram, ward

# Load the data
url = 'https://raw.githubusercontent.com/HiwaTase/Machine-Learning/main/healthINS_balanced.csv'
df = pd.read_csv(url)

# Reduce the dataset size by sampling
df_sampled = df.sample(n=10000, random_state=42)

# Proceed with the same preprocessing steps
X = df_sampled.drop('nohealthins', axis=1)
y = df_sampled['nohealthins']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Define the preprocessing steps for numerical and categorical data
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Apply the preprocessing steps to the training and test data
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Convert the sparse matrix to a dense matrix
X_train_dense = X_train_processed.toarray()
X_test_dense = X_test_processed.toarray()
```

```python
# Apply regular PCA to determine the number of components
pca = PCA(n_components=0.95)
pca.fit(X_train_dense)
n_components = pca.n_components_

print(f"Number of components to retain 95% of the variance: {n_components}")

# Apply Incremental PCA to the training data with the determined number of components
ipca = IncrementalPCA(n_components=n_components, batch_size=200)
X_train_ipca = ipca.fit_transform(X_train_dense)

# Plot the cumulative explained variance
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(ipca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance Explained')
plt.title('Explained Variance vs. Number of Components')
plt.grid()
plt.show()

# Transform the test set using the same Incremental PCA transformation
X_test_ipca = ipca.transform(X_test_dense)
```

```python
# Function to plot the elbow method for k-Means
def plot_elbow(X):
    distortions = []
    K = range(1, 11)
    for k in K:
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(X)
        distortions.append(kmeanModel.inertia_)
    plt.figure(figsize=(8, 5))
    plt.plot(K, distortions, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Distortion')
    plt.title('The Elbow Method showing the optimal k')
    plt.grid()
    plt.show()

# Plot elbow method for k-Means
plot_elbow(X_train_dense)

# k-Means Clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_train_dense)
y_kmeans = kmeans.labels_
kmeans_ari = adjusted_rand_score(y_train, y_kmeans)
kmeans_silhouette = silhouette_score(X_train_dense, y_kmeans)

mglearn.discrete_scatter(X_train_dense[:, 0], X_train_dense[:, 1], y_kmeans)
plt.legend(["Cluster 0", "Cluster 1", "Cluster 2"], loc="best")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.title("k-Means Clustering without PCA")
plt.show()

# Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=3)
y_agglo = agglo.fit_predict(X_train_dense)
agglo_ari = adjusted_rand_score(y_train, y_agglo)
agglo_silhouette = silhouette_score(X_train_dense, y_agglo)

mglearn.discrete_scatter(X_train_dense[:, 0], X_train_dense[:, 1], y_agglo)
plt.legend(["Cluster 0", "Cluster 1", "Cluster 2"], loc="best")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.title("Agglomerative Clustering without PCA")
plt.show()

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X_train_dense)
# Filter out noise points (-1) for silhouette score
mask = y_dbscan != -1
dbscan_ari = adjusted_rand_score(y_train, y_dbscan)
dbscan_silhouette = silhouette_score(X_train_dense[mask], y_dbscan[mask])

mglearn.discrete_scatter(X_train_dense[:, 0], X_train_dense[:, 1], y_dbscan)
plt.legend(["Cluster 0", "Cluster 1", "Cluster 2"], loc="best")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.title("DBSCAN Clustering without PCA")
plt.show()

print("Clustering results without PCA:")
print(f"k-Means ARI: {kmeans_ari}, Silhouette: {kmeans_silhouette}")
print(f"Agglomerative Clustering ARI: {agglo_ari}, Silhouette: {agglo_silhouette}")
print(f"DBSCAN ARI: {dbscan_ari}, Silhouette: {dbscan_silhouette}")
```
